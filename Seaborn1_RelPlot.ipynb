```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px #for datasets
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips, x ='total_bill', y='tip')
```

```python
sns.scatterplot(data=tips)
```

```python
sns.relplot(data=tips, x = 'total_bill', y='tip', kind='scatter')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='line')
```

```python
sns.relplot(data=tips, x = 'total_bill', y='tip', kind='scatter', hue='sex')
```

```python
sns.scatterplot(data=tips, x ='total_bill', y='tip',hue='sex')
```

```python
sns.scatterplot(data=tips, x ='total_bill', y='tip',hue='sex',style='time')
```

```python
sns.scatterplot(data=tips, x ='total_bill', y='tip',hue='sex',style='time',size='size')
```

```python
#another dataset
#dataset present in data of plotly
gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country']=='India']
temp_df
```

```python
sns.lineplot(data=temp_df, x='year', y='lifeExp') #axes function
```

```python
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line') #figure function
```

```python
temp2_df = gap[gap['country'].isin(['India', 'Pakistan', 'China'])]
temp2_df
```

```python
sns.relplot(kind='line', data=temp2_df, x='year', y='lifeExp', hue='country')
```

```python
sns.lineplot( data=temp2_df, x='year', y='lifeExp', hue='country')
```

```python
temp3_df = gap[gap['country'].isin(['India', 'Brazil', 'Germany'])]
temp3_df
```

```python
sns.relplot(kind='line', data=temp3_df, x='year', y='lifeExp', hue='country', style='continent',size='continent')
```
