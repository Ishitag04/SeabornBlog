```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as ps
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(kind='scatter', data=tips, x = 'total_bill', y='tip')
```

```python
sns.relplot(kind='scatter', data=tips, x = 'total_bill', y='tip', hue='sex')
```

```python
sns.relplot(kind='scatter', data=tips, x = 'total_bill', y='tip', col='sex')
#sex column me jitni categories h utne scatter plot bna do side by side
```

```python
sns.relplot(kind='scatter', data=tips, x = 'total_bill', y='tip', row='sex')
```

```python
sns.relplot(kind='scatter', data=tips, x = 'total_bill', y='tip', row='smoker', col='sex')
```

```python
sns.relplot(kind='scatter', data=tips, x = 'total_bill', y='tip', col='sex', row='day')
```

```python
sns.relplot(kind='line', data=tips, x='total_bill', y='tip', row='sex', col='smoker')
```

```python
sns.lineplot(data=tips, x = 'total_bill', y='tip', col='sex', row='smoker')
#facet plot will work only with figure level functions
```
