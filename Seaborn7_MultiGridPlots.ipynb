```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# figure level function: relplot, displot, catplot, lmplot
```

```python
# Facetgrid
sns.catplot(data=tips, x='sex', y='total_bill', kind='violin', col='day', row='time')
```

```python
# One more way of doing the above thing is using the facetgrid
# facetgrid gives more flexibiliy
```

```python
g = sns.FacetGrid(data=tips, col='day', row='time')
g.map(sns.violinplot, 'sex', 'total_bill')
```

```python
# Pair plot
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

```python
# PairGrid
```

```python
g = sns.PairGrid(data=iris)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris)
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris)
g.map_diag(sns.countplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris)
g.map_diag(sns.boxplot)
g.map_offdiag(sns.histplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.kdeplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species', vars=['sepal_length', 'sepal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
