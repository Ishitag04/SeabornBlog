```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
tips
```

```python
sns.stripplot(data=tips, x='day',y='total_bill')
```

```python
sns.stripplot(data=tips, x='day',y='total_bill', jitter=False)
```

```python
sns.stripplot(data=tips, x='day',y='total_bill', jitter=False, hue='day')
```

```python
#figure level function
sns.catplot(data=tips, x='day', y='total_bill', kind='strip')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='strip', jitter=0.3)
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='strip', hue='sex')
```

```python
#Swarm plot
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill',hue='sex')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
sns.boxplot(data=tips, y='total_bill')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box',hue='sex')
```
